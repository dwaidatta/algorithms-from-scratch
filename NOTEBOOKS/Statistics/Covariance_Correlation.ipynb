## Importing the Libraries

```python
import numpy as np 
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_regression

X, Y = make_regression(n_samples=1000, n_features=100, n_targets=100, noise=10, random_state=42)
# Y, _ = make_regression(n_samples=1000, n_features=10, n_targets=1, noise=10, random_state=24)
X1, X2, X3, X4 = X[:, 0], X[:, 1], X[:, 2], X[:, 3]
Y1, Y2, Y3, Y4 = Y[:, 0], Y[:, 1], Y[:, 2], Y[:, 3]
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(X1, Y4)
```

## Covariance And Correlation

```python
def Cov(X, Y):
    mean_x = np.mean(X)
    mean_y = np.mean(Y)
    cov = np.sum((X - mean_x) * (Y - mean_y)) / (len(X))
    return cov

def Corr(X, Y):
    corr = Cov(X, Y) / (np.std(X) * np.std(Y))
    return corr

Corr(X[:, 0], Y[:, 0])
```

```python
correlation_matrix = np.zeros((100, 100))
covariance_matrix = np.zeros((100, 100))

for i in range(100):
    for j in range(100):
        correlation_matrix[i][j] = Corr(X[:, i], Y[:, j])
        
for i in range(100):
    for j in range(100):
        covariance_matrix[i][j] = Cov(X[:, i], Y[:, j])


correlation_matrix, covariance_matrix
```

```python
x_labels = [f"Y{j+1}" for j in range(100)]
y_labels = [f"X{i+1}" for i in range(100)]

# Create a heatmap
plt.figure(figsize=(24, 20))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm",
            xticklabels=x_labels, yticklabels=y_labels, cbar_kws={'label': 'Correlation'})

plt.title("Correlation Matrix between X and Y Features", fontsize=14)
plt.xlabel("Y Features", fontsize=12)
plt.ylabel("X Features", fontsize=12)
plt.tight_layout()
plt.show()
```
